```python
flatten_data = [] #to collect input
target_data = []  #to to collect output
datadir = '/content/drive/MyDrive/Data'
categories=['Dog','Cat'] #as per the listed directory we need to specify the dog and cat in this list
```

```python
for i in categories: #i=Dog
  print("Loading Categories")
  path=os.path.join(datadir,i)
  for img in os.listdir(path):
    img_arr = imread(os.path.join(path,img))
    img_resize=resize(img_arr,(150,150,3))
    flatten_data.append(img_resize.flatten())
    target_data.append(categories.index(i))

  print("Loaded Completed")
```

```python
flat_data = np.array(flatten_data)
target_data = np.array(target_data)
#150*150*3 = 67500 cols
```

```python
df=pd.DataFrame(flat_data)
df
```

```python
df['target']=target_data
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
acc = accuracy_score(y_test,y_pred)
acc
```
