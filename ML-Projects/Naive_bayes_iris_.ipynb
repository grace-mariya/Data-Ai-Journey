```python
x = df.drop(columns=['Species','Id'])
x.ndim
y = df['Species']
y.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

In this algorithm it is not necessary to do normalization(StandardScaler) , because naive bayes is working on the basis of probability..(yes/no or 1 or 0).so therefore we dont need to scale it

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
print(gnb.predict([[	6.7,	3.0,	5.2,	2.3]]))
```

```python
y_pred = gnb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

```

```python
y_train
```

```python
labels = ['Setosa','Versicolor','Virginica']
cmd = ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot(cmap=plt.cm.Blues)
```
