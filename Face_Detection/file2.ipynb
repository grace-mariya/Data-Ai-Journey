```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('group3.jpg')
plt.imshow(img)
```

```python
img_conv = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_conv)
```

```python
face_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_data
```

```python
face=face_data.detectMultiScale(img_conv,minNeighbors=3)
face
```

```python
for x,y,w,h in face:
    cv2.rectangle(img_conv,(x,y),(x+w,y+h),color=(0,0,255),thickness=2)
```
