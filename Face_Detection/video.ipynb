```python
import cv2
```

```python
faces=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
video = cv2.VideoCapture(0) #in VideoCapture there is read()
while True:
    suc,img = video.read()  #image stored in img - it is in BGR format
    gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #for more better feature extraction we convert to gray scale
    face = faces.detectMultiScale(gray_img,minNeighbors=5)
    for x,y,w,h in face:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4) #but for drawing rectangle we pass the original image itself which is in BGR format
    cv2.imshow('video',img)
    if cv2.waitKey(1)&0XFF == ord('q'): #it has to display img untill pressing 'q', after pressing 'q' close the frame
        break
video.release()             # close the loop (release the wecan resource)
cv2.destroyAllWindows()     # close all opencv windows
```
