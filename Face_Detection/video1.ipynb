```python
import cv2
```

```python
faces=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
video=cv2.VideoCapture(r'C:\Users\Grace Mariya\Deep_Learning\Face_Detection\videofile.mp4')
while True:
    suc,img=video.read()
    if suc:
        gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

        face = faces.detectMultiScale(gray_img,minNeighbors=5)
        for x,y,w,h in face:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4) 
        cv2.imshow('video',img)
        if cv2.waitKey(1)&0XFF == ord('q'): 
            break
    else:
        print("Video Finished")
        break
video.release()             
cv2.destroyAllWindows()  
```
