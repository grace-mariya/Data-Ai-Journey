```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images (1).jpg")
plt.imshow(img)
```

```python
conv_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_data=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_data
```

To detect face

```python
face=face_data.detectMultiScale(conv_img)#,minNeighbors = 5)
face
```

To draw rectangle: only left top is only available:

```python
#rectangle:
for x,y,w,h in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),3)
```
