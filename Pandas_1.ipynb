```python
import pandas as pd
data=pd.read_excel('/content/drive/MyDrive/data/Sample-Superstore-Subset-Excel (2).xlsx')
print(data.info())
print("/n")
print(data.describe)
print("/n")
print(data.head(20))
```

```python
#Data Cleaning
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
print(data['Order Date'])
print(data['Ship Date'])
print(data.head(20))
```

```python
#Verifying the data
#check for missing values
data.isnull().sum()

```

```python
#check for duplicates
chk_duplicate=data.duplicated().sum()
print(chk_duplicate)
```

After cleaning:

```python
# Summary statistics for numerical columns
numerical_summary = data.describe()
print("Summary Statistics for Numerical Columns:")
display(numerical_summary)

# Value counts for categorical columns
print("\nValue Counts for Categorical Columns:")
for column in data.select_dtypes(include='object').columns:
    print(f"\nValue counts for column: {column}")
    display(data[column].value_counts())
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Category:")
display(sales_by_category)
```

```python
sales_by_region = data.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Region:")
display(sales_by_region)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Discount', y='Profit')
plt.title('Relationship between Discount and Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
sales_by_customer = data.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)
print("\nTop 10 Customers by Total Sales:")
display(sales_by_customer.head(10))
```

```python
profit_by_customer = data.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False)
print("\nTop 10 Customers by Total Profit:")
display(profit_by_customer.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
data[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle('Distribution of Numerical Variables', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```
