```python
import cv2
import pytesseract
```

```python
img = cv2.imread('resume2.jpeg')
img
```

```python
cv2.imshow('Frame1',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
txt = pytesseract.image_to_string(img)
print(txt)
```

```python
dic = pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
import re
```

```python
n=len(dic['text'])
n
```

```python
for i in range(n):
    if re.search(r'[a-z0-9]+@[a-z]+.[a-z]',dic['text'][i]):
        print(dic['text'][i])
        x=dic['left'][i]
        y=dic['top'][i]
        w=dic['width'][i]
        h=dic['height'][i]

        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)
cv2.imshow('Frame1',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
for i in range(n):
    if re.search(r'\s?[0-9]\d{9}',dic['text'][i]):
        print(dic['text'][i])
        x=dic['left'][i]
        y=dic['top'][i]
        w=dic['width'][i]
        h=dic['height'][i]

        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)
cv2.imshow('Frame2',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
lst = [i.lower() for i in lst]

n = len(dic['text'])
count = 0

for i in range(n):
    text = dic['text'][i].strip().lower()
    
    if text != " " and text in lst:
        count += 1
        print(text)
        
        x = dic['left'][i]
        y = dic['top'][i]
        w = dic['width'][i]
        h = dic['height'][i]
        
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('Frame 3',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
