```python
import cv2
import pytesseract
```

Read image icr1.jpeg

```python
img = cv2.imread('ocr1.jpeg')
cv2.imshow('Frame',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

Setting tesseract path , for windows

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

Extract text from img

```python
text = pytesseract.image_to_string(img)
print(text) #print text , otherwise it will form \n
```

Text Properties of the extracted text:
we must need to know the properties of extracted text
we will store them in a dictionary as a key value pair

```python
dic = pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```
