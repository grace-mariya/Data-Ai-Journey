```python
df=df.drop(columns=['random_notes','noise_col'])
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Glucose']=df['Glucose'].fillna(df['Glucose'].mean())
df['Blood Pressure']=df['Blood Pressure'].fillna(df['Blood Pressure'].mean())
df=df.dropna(subset=['Medical Condition'])
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
```

```python
df=df.copy()
df['Gender']=lab.fit_transform(df['Gender'])
df['Medical Condition']=lab.fit_transform(df['Medical Condition'])
df
```

```python
x=df.drop(columns=['Medical Condition'])
y=df['Medical Condition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_res,y_train_res = smote.fit_resample(x_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train_res)
x_train_res = scaler.transform(x_train_res)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=7,weights='distance')
knn.fit(x_train_res,y_train_res)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
y_train.head(20)
```

```python
labels=['Diabetes','Healthy','Obesity','Arthritis','Hypertension','Asthma','Cancer']
fig,ax= plt.subplots(figsize=(10, 8))
cmd = ConfusionMatrixDisplay(cm, display_labels=labels)
cmd.plot(ax=ax,cmap='cubehelix_r')


plt.show()
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred,target_names=labels))
```

```python
y_prob_knn = knn.predict_proba(x_test)
y_prob_knn[:5]
```

```python
from sklearn.metrics import roc_curve, auc

n_classes = len(labels)

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_prob_knn[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'ROC curve (Class {labels[i]}) = {roc_auc[i]:.2f}')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title('Receiver Operating Characteristic (ROC) Curve - KNN')
plt.legend(loc="lower right")
plt.show()
```
