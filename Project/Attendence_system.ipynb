```python
import cv2
import numpy as np
import face_recognition as fr
import os
```

```python
path = 'Images'
path
```

```python
mylist = os.listdir(path)
mylist
```

```python
imgs = []
ImgNames=[]

for i in mylist:
    img_path=os.path.join(path,i)
    current_img = cv2.imread(img_path)
    imgs.append(current_img)
    ImgNames.append(i.split('.')[0])
```

```python
def faceencodings(imgs):
    encodelist=[]
    for img in imgs:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img)
        encode = fr.face_encodings(img,face_in_frame)[0]
        encodelist.append(encode)
    return encodelist
```

```python
encodelist_knownface = faceencodings(imgs)
print(encodelist_knownface)
```

```python
import csv
from datetime import datetime
```

```python
video = cv2.VideoCapture(0)
logged_names = set()

file_path = 'attendance.csv'
file_exists = os.path.isfile(file_path)

with open(file_path, 'a', newline='', buffering=1) as f:   # ✅ buffering added
    writer = csv.writer(f)

    if not file_exists:
        writer.writerow(["NAME", "TIME"])
        f.flush()

    while True:
        suc, img = video.read()

        if not suc:
            print("Camera not working")
            break

        img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img1)
        face_in_location = fr.face_encodings(img1, face_in_frame)

        print("Faces detected:", len(face_in_frame))   # ✅ DEBUG

        for face_enc, face_loc in zip(face_in_location, face_in_frame):

            if len(encodelist_knownface) == 0:
                print("No known faces loaded")
                continue

            distance = fr.face_distance(encodelist_knownface, face_enc)
            matchindex = np.argmin(distance)

            print("Distance:", distance[matchindex])   # ✅ DEBUG

            # ✅ BETTER recognition logic
            if distance[matchindex] < 0.5:
                name = ImgNames[matchindex]
            else:
                name = "unknown"

            print("Detected:", name)   # ✅ DEBUG

            y1, x2, y2, x1 = face_loc
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(img, name, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)

            # ✅ WRITE BLOCK
            if name != "unknown" and name not in logged_names:
                now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

                print("WRITING TO CSV:", name, now)   # 🔥 CRITICAL DEBUG

                writer.writerow([name, now])
                f.flush()   # ✅ MUST

                logged_names.add(name)

        cv2.imshow('Video', img)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

video.release()
cv2.destroyAllWindows()
```
