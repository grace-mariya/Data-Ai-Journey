```python
import cv2
import numpy as np
import os
import face_recognition as fr
```

```python
path = 'Images'
path
```

```python
os.listdir(path)
```

```python
mylist = os.listdir(path)
mylist
```

```python

imgs=[]
classnames=[]

for i in mylist:
    imgpath = os.path.join(path,i)#setting img path
    current_img = cv2.imread(imgpath)#reading each img 
    imgs.append(current_img) #appending each img 
    classnames.append(i.split('.')[0]) #appending names without '.jpg ' so we are splitting the name and taking 0th index 
```

```python
def faceencodings(imgs):
    encodelist=[]
    for img in imgs:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img)
        print(face_in_frame)
        encode = fr.face_encodings(img,face_in_frame)[0]
        encodelist.append(encode)
    return encodelist
```

```python
encodelist_knownface = faceencodings(imgs)
print(encodelist_knownface)
```

```python


video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()
    if suc:
        img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img1)
        face_in_location = fr.face_encodings(img1, face_in_frame)

        for enc_fac, face_loc in zip(face_in_location, face_in_frame):
            matches = fr.compare_faces(encodelist_knownface, enc_fac)
            face_dis = fr.face_distance(encodelist_knownface, enc_fac)
            matchindex = np.argmin(face_dis)

            if matches[matchindex]:
                name = classnames[matchindex]
            else:
                name = "Unknown"

            # unpack coordinates correctly
            y1, x2, y2, x1 = face_loc
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(img, name, (x1, y1 - 10), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)

        # show the image with rectangles/text
        cv2.imshow('video', img)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        print("Video completed")
        break

video.release()
cv2.destroyAllWindows()
```
