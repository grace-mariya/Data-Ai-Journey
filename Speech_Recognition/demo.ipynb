```python
#text to voice
#install library -> pyttsx3

import pyttsx3 #importing library
txt_sp = pyttsx3.init() #define a class init()
voice=txt_sp.GetProperty('voices')
txt_sp.setProperty('voice',voice[1].id)
text = input("Enter the Text:")#from user we get a text
txt_sp.say(text)#text to voice
txt_sp.runAndWait()#until it converts to voice it has to run and wait

```
