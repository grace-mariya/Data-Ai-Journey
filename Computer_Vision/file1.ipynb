To convert to grey scale

```python
img1=cv2.imread('download.jpg',cv2.IMREAD_GRAYSCALE)
img1
```

```python
img1.shape
```

```python
cv2.imshow('Frame2',img1)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

To save image

```python
cv2.imwrite('cat_image.jpg',img1)
```

To convert into binary:
rgb -> binary is not possible..
grey scale - > binary is only possible

```python
thresh,bin_img = cv2.threshold(img1,125,255,cv2.THRESH_BINARY)
bin_img
```

```python
bin_img.shape
```

```python
cv2.imshow('Frame3',bin_img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
