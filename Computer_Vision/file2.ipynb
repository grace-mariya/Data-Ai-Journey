```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images.jpg')
plt.imshow(img)
```

```python
conv_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

Bounding BOX- we can create or draw rectangles or circles etc..in an image or texts in an image

#Rectangle:
#2 points:
#-->Left Top
#-->Right Bottom

```python
cv2.rectangle(conv_img,pt1=(70,5),pt2=(240,140),color=(0,0,255),thickness=3)
plt.imshow(conv_img)
```

```python
img2 = cv2.imread('download.jpg')
plt.imshow(img2)
```

```python
conv_img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img2)
```

Circle:
2 points:
--> center
-->radius

```python
cv2.circle(conv_img2,center=(120,100),radius=50,color=(0,0,255),thickness=3)
plt.imshow(conv_img2)
```

To add text in an image:

```python
cv2.putText(conv_img2,text='CUTE CAT',org=(40,35),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.5,color=(255,0,0),thickness=1)
plt.imshow(conv_img2)
```
