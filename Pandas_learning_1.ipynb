```python
#fill empty columns with a value
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/data/data (1).csv')

df.fillna(130,inplace = True)

print(df.to_string())
```

```python
#To fill specific column:
df.fillna({"Calories": 130}, inplace=True)
print(df.to_string())
```

```python
#fill with mean median mode value:

#Mean
x=df["Calories"].mean()
df.fillna({"Calories":x}, inplace=True)
print(df.to_string())
```

```python
#Median
x = df["Calories"].median()

df.fillna({"Calories": x}, inplace=True)
print(df.to_string())
```

```python
#Mode
x = df["Calories"].mode()[0]

df.fillna({"Calories": x}, inplace=True)
print(df.to_string())
```

```python
data = [
    ["Alice", 20, 85],
    ["Bob", 21, 90],
    ["Charlie", 19, 88],
    ["David", 22, 92]
]

df = pd.DataFrame(data, columns=["Name", "Age", "Marks"])
print(df)
```

```python
print(df.loc[0])      # by label/index
print(df.iloc[1])     # by position
```

```python
print(df.loc[0:2, ["Name", "Marks"]])   # rows 0–2, columns Name & Marks
print("\n")
print(df.iloc[1:3, 0:2])                # rows 1–2, first 2 columns
```

```python
#Add
df["Grade"]=['A','A','B','A']
print(df)
```

```python
#rename
df = df.rename(columns={"Marks": "Score"})
print(df)
```

```python
#remove
df=df.drop("Age",axis=1)
print(df)
```

```python
students = pd.DataFrame({
    "Student": ["John", "Mia", "Alex", "Sara"],
    "Math": [85, 78, 92, 88],
    "Science": [89, 74, 95, 80],
    "English": [90, 82, 85, 87]
})
```

```python
#Display
print(students)
```

```python
#Display only math and english col
print(students[['Math','English']])
```

```python
print(students.iloc[1:3])
```

```python
#Add col
students['Average']=students[["Math", "Science", "English"]].mean(axis=1)
print(students)
```

```python
#Drop Science col
students = students.drop("Science", axis=1)
print(students)
```

```python
students = students.rename(columns={"English":"Eng"})
print(students)
```
