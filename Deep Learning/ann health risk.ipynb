```python
df['Physical Activity'] = df['Physical Activity'].apply(lambda x: 0 if x < 0 else x)
```

```python
df['Sleep Hours']=df['Sleep Hours'].round().astype(int)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Glucose']=df['Glucose'].fillna(df['Glucose'].mean())
df['Blood Pressure']=df['Blood Pressure'].fillna(df['Blood Pressure'].mean())
df['Physical Activity']=df['Physical Activity'].fillna(df['Physical Activity'].mean())
df=df.dropna(subset=['Medical Condition'])
df
```

```python
df.drop(['random_notes','noise_col'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender'] = lab.fit_transform(df['Gender'])
df['Medical Condition'] = lab.fit_transform(df['Medical Condition'])
```

```python
df.dtypes
```

```python
x=df.drop(['Medical Condition'], axis=1)
y=df['Medical Condition']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_res,y_train_res = smote.fit_resample(x_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train_res)
x_train_res = scaler.transform(x_train_res)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann=Sequential()
```

```python
ann.add(Dense(17,activation='relu'))
ann.add(Dense(16,activation='relu'))
ann.add(Dense(8,activation='relu'))
ann.add(Dense(4,activation='relu'))
ann.add(Dense(7,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train_res,y_train_res,epochs=30,batch_size=30)
```

```python
ann.evaluate(x_test,y_test)
```
