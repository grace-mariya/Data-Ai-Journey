# Day 5

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [25, 32, 28, 41, 19],
    "Marks": [82, 75, 90, 66, 95],
    "city": ["Hong Kong", "Manchester", "Vienna" ,"Banglore", "Adelaide"]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Apply conditions
print(df[df["Age"]>30])
print(df[(df["Age"]>25) & (df["Age"]>80)])
```

```python
#Use .loc[] and .iloc[] with Conditions

#.loc[] → label-based
#.iloc[] → position-based

# Using loc
df.loc[df["Marks"] > 80, ["Name", "Marks"]]   # select rows where Marks > 80

# Using iloc (row/column numbers)
df.iloc[0:3, 0:2]   # first 3 rows, first 2 columns
```

```python
#Use the query method
df.query("Marks > 80 and Age < 30")
```

```python
df.query("Marks > 80")
```

```python
df[(df["Age"]>20) & (df["Age"]<30) ]
```

```python
print(df.query("city  == 'Adelaide'"))
```

```python
print(df["Marks"].mean())
print(df["Marks"].sum())
print(df["Marks"].std())
print(df.describe())
```

```python
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])
print(numeric_df.corr())
```

```python
#Apply math using Numpy
import numpy as np
df["Bonus"]=np.sqrt(df["Marks"])*2
print(df)
```

```python
#Apply custom functions
def grade(marks):
  if marks >= 90:
    return "A"
  elif marks >=75:
    return "B"
  else:
    return "C"

df["Grade"] = df["Marks"].apply(grade)
print(df)
```

Find:

Average marks

Correlation between Age and Marks

Create a new column Grade using a function.

Add a new column Bonus = Marks * 0.05.

```python
import pandas as pd

data=pd.DataFrame({
    "Name": ["Grace", "Flerin", "Joel", "Awin", "Treasa"],
    "Age": [22, 19, 23, 23, 21 ],
    "Marks":[85, 73, 97, 80, 76],
    "City": ["Kochi", "Banglore", "Adelaide", "Thrissur", "Vienna"]

})

print(data)
print("--------------------------------")
print("Average Marks:",data["Marks"].mean())
print("--------------------------------")
print("Correlation between Age and marks:")
new_data=data.select_dtypes(include=["number"])
print(new_data.corr())
print("--------------------------------")
print("Show Grade:")

def grade(marks):
  if marks>90:
    return "S"
  elif marks>80:
    return "A"
  elif marks >70:
    return "B"
  else:
    return "C"

data["Grade"]=data["Marks"].apply(grade)
print(data)

print("--------------------------------")
print("Add Bonus:")
data["Bonus"]=data["Marks"]*0.05
print(data)
```
